```python
%time

import numpy as np

f = open("/Users/shu-wei/Desktop/GW150914/fig1_residual_H.txt")
list_t = []
list_hstrain = []
for line in f.readlines():
#     print(line)
#     print(line[0])
#     print("######")
    s = line.split(' ')
#     print(len(s))
#     print(s)
    list_t.append(s[0])
    list_hstrain.append(s[1].rstrip())
    
f.close()

list_t = np.array(list_t[1:]).astype(np.float)
list_hstrain = np.array(list_hstrain[1:]).astype(np.float)

# np.savez_compressed("", T = list_t, Strain = list_strain, information=[""])
```

```python
# file = np.load("")
# file.files

# list_t = file["T"]
# list_strain = file["Strain"]

import matplotlib.pyplot as plt

plt.plot(list_t[:], list_hstrain[:])
plt.xlabel('GPS Time (s)')
plt.ylabel('H1 Strain, 4096 HZ, GW150914')

plt.show()
```
