```python
%time

import numpy as np

f = open("/Users/shu-wei/Desktop/GW150914/H1_GDS_CALIB_STRAIN.txt")
list_f = []
list_hasd = []
for line in f.readlines():
#     print(line)
#     print(line[0])
#     print("######")
    s = line.split(' ')
#     print(len(s))
#     print(s)
    list_f.append(s[0])
    list_hasd.append(s[1].rstrip())
    
f.close()

list_f = np.array(list_f[1:]).astype(np.float)
list_hasd = np.array(list_hasd[1:]).astype(np.float)

# np.savez_compressed("", T = list_t, Strain = list_strain, information=[""])
```

```python
# file = np.load("")
# file.files

# list_t = file["T"]
# list_strain = file["Strain"]

%matplotlib inline
%config InlineBackend.figure_format = 'retina'
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab

plt.figure(figsize=(10,8))
plt.loglog(list_f[:], np.sqrt(list_hasd[:]), 'b', label='H1 strain')

plt.xlabel('Freq (Hz)')
plt.ylabel('ASD (strain/rtHz)')
plt.show()
```
